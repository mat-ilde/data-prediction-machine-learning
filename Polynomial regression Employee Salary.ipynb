```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
Salary=pd.read_csv('EmployeeSalary.csv')
Salary
```

```python
sns.jointplot(x='Years of Experience', y='Salary', data=Salary)
```

```python
x=Salary[['Years of Experience']]
y=Salary[['Salary']]
```

```python
#in this case i'll use all the data for trainning
x_train=x
y_train=y
```

# solution using linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression(fit_intercept=True)
```

```python
regressor.fit(x_train,y_train)
```

```python
print('Linear model coeff(m)',regressor.coef_)
print('Linear model intercept(b)',regressor.intercept_)
```

```python
plt.scatter(x_train,y_train, color='red')
plt.plot(x_train.values,regressor.predict(x_train), color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary/year $')
plt.title('Years of Experience vs. Salary @ (Linear)')
```

# solution using polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_regressor=PolynomialFeatures(5)#degree desviation
```

```python
x_colums=poly_regressor.fit_transform(x_train)#multiple arrays
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_colums,y_train)
```

```python
y_predict=regressor.predict(poly_regressor.fit_transform(x_train))
```

```python
plt.scatter(x_train,y_train, color='red')
plt.plot(x_train.values,y_predict, color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary/year $')
plt.title('Years of Experience vs. Salary @ (Polynomial)')
```
