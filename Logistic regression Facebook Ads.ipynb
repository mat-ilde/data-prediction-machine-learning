```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
training_set=pd.read_csv('FacebookAds.csv',encoding='ISO-8859-1')
```

```python
training_set
```

```python
click=training_set[training_set['Clicked']==1]
no_click=training_set[training_set['Clicked']==0]
```

```python
print("Total", len(training_set))
print("Number of users who clicked", len(click))
print("Number of users who did not clicke", len(no_click))
```

```python
sns.scatterplot(x=training_set['Time Spent on Site'], y=training_set['Salary'], hue = training_set['Clicked'])
```
