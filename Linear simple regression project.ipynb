```python
import numpy as np #data analysis
import matplotlib.pyplot as plt #data visualization
import seaborn as sns #nicer data visualization
import pandas as pd #mathematical operations on arrays
```

```python
#data
IceCream=pd.read_csv('IceCreamData.csv')
```

```python
sns.pairplot(IceCream)
```

```python
#creating independent and dependent variable
x=IceCream[['Temperature']]#must be 2d
y=IceCream['Revenue']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#test_size=33 data to test, random_state=42 to get the same result
#X_train is a dataframe, we must to access to the data value (x.values)
X_train, X_test, y_train, y_test=train_test_split(x.values, y, test_size=33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
#the model
regressor=LinearRegression(fit_intercept=True)
regressor.fit(X_train, y_train)
```

```python
plt.scatter(X_train,y_train, color='black')
plt.plot(X_train, regressor.predict(X_train),color='red')

plt.ylabel('Revenue[Dollars]')
plt.xlabel('Revenue[DegC]')
plt.title('Revenue Generated vs. Temperature @ Ice Cream Stand (Training set)')
```

```python
X_train.shape
```

```python
print('Linear Model Coefficient(m)',regressor.coef_)#slope
print('Linear Model Coefficient(b)',regressor.intercept_)# Y intercept
```

```python
#testing
y_predict=regressor.predict(X_test)
```
