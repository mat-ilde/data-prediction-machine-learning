```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cost=pd.read_csv('EconomiesOfScale.csv')
```

```python
sns.jointplot(x='Number of Units',y='Manufacturing Cost',data=cost)
```

```python
cost.describe()
```

```python
x=cost[['Number of Units']]
y=cost['Manufacturing Cost']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x.values, y, test_size=33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression(fit_intercept=True)
```

```python
regressor.fit(X_train,y_train)
```

```python
print('Linear Model Coefficient(m)',regressor.coef_)#slope
print('Linear Model Coefficient(b)',regressor.intercept_)# Y intercept
```

```python
plt.scatter(X_train,y_train, color='blue')
plt.plot(X_train,regressor.predict(X_train), color='red')

x=cost[['Number of Units']]
y=cost['Manufacturing Cost']

plt.xlabel('Number of Units')
plt.ylabel('Manufacturing Cost[in millions]')
plt.title('Number of Units vs. Manufacturing Cost @ (Linear)')
```

```python
x_train=x
y_train=y
```

# Polynomial solution

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_regressor=PolynomialFeatures(5)#degree desviation
```

```python
x_colums=poly_regressor.fit_transform(x_train)#multiple arrays
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_colums,y_train)
```

```python
y_predict=regressor.predict(poly_regressor.fit_transform(x_train))
```

```python
plt.scatter(x_train,y_train, color='red')
plt.plot(x_train.values,y_predict, color='blue')
plt.xlabel('Number of Units')
plt.ylabel('Manufacturing Cost[in millions]')
plt.title('Number of Units vs. Manufacturing Cost @ (Polynomial)')
```
